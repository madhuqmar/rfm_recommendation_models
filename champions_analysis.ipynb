```python
import pandas as pd
```

```python
champions_df = pd.read_csv("outputs/champions.csv")
champions_df
```

```python
df3 = pd.read_csv("data/Ticket_Product_Details_14Nov23_4pm.csv", encoding='ISO-8859-1', low_memory=False)
df3.head()
```

```python
client_services = df3[['TicketID', 'Type', 'ProdID', 'Descr', 'Qty', 'Retail_Price', 'Start_Time']]
client_services
```

```python
df4 = pd.read_csv("data/Tickets_14Nov23_4pm.csv", encoding='ISO-8859-1', low_memory=False)
df4.head()
```

```python
df4_clients = df4[['TicketID', 'ClientID']]
df4_clients
```

```python
# Try to convert the 'column_name' to int64, handling errors
try:
    df4['TicketID'] = df4['TicketID'].astype('int64')
except ValueError:
    # Handle non-numeric values, for example, replacing them with NaN
    df4['TicketID'] = pd.to_numeric(df4['TicketID'], errors='coerce')

#print(df4.dtypes)
```

```python
df4_clients = df4[['TicketID', 'ClientID']]
df4_clients
```

```python
services_df = pd.merge(client_services, df4_clients, on='TicketID', how='inner')
services_df
```

```python
champions_services = services_df[services_df['ClientID'].isin(champions_df['ClientID'])]
champions_services
```

```python
# Assuming you have a DataFrame named sales_df
services_counts = champions_services['Descr'].value_counts()
```

```python
services_counts = services_counts[services_counts.values > 1]
len(services_counts)
```

```python
champ_service_counts = pd.DataFrame(services_counts)
champ_service_counts
```

```python
services_costs = services_df[['ProdID', 'Descr', 'Retail_Price']]
services_costs
```

```python
services_costs = services_costs.drop_duplicates()
len(services_costs)
```

```python
champ_service_costs = pd.merge(champ_service_counts, services_costs, on='Descr', how='left')
champ_service_costs
```

```python
len(champ_service_costs['ProdID'].unique())
```

```python
champ_service_costs = champ_service_costs.groupby('ProdID').agg({'Retail_Price': 'mean',
                                                                        'Descr': 'first'}).reset_index()
champ_service_costs
```

```python
champ_service_costs = champ_service_costs.sort_values(by='Retail_Price', ascending=False)
champ_service_costs
```
